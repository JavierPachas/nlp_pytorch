## **Corpora, Tokens and Types**

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
text = "Mary, don't slap the green witch"

print([str(token) for token in nlp(text.lower())])
```

```python
from nltk.tokenize import TweetTokenizer

tweet = "Snow White and the Seven Degrees #MakeAMovieCold@midnight:-)"
tokenizer = TweetTokenizer()
print(tokenizer.tokenize(tweet.lower()))
```

## **Unigrams, Bigrams, Trigrams, ... N-grams**

```python
def n_gram(text,n):
    #takes tokens or text, returns a list of n-grams
    return [text[i:i+n] for i in range(len(text)-n+1)]
```

```python
cleaned = ['mary', ',', 'do', "n't", 'slap', 'the', 'green', 'witch']
print(n_gram(cleaned, 3))
```

```python
#Lemmatization: reducing words to their root form
doc = nlp(u'he was running late')
for token in doc:
    print('{} - {}'.format(token, token.lemma_))
```

**Categorizing words: Part of Speech (POS) Tagging**

```python
doc = nlp(u'Mary slapped the green witch.')
for token in doc:
    print('{} - {}'.format(token, token.pos_))
```

**Categorizing spans: Chunking and named entity recognition**

```python
for chunk in doc.noun_chunks:
    print('{} - {}'.format(chunk, chunk.label_))
```
